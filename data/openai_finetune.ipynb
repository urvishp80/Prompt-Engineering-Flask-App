```python
# COMPLETIONS_MODEL = "ada:ft-hyly-ai:greystar-2023-02-14-16-49-00"
COMPLETIONS_MODEL = "text-ada-001"
EMBEDDING_MODEL = "text-embedding-ada-002"

MAX_SECTION_LEN = 1000
SEPARATOR = "\n* "
ENCODING = "cl100k_base"  # encoding for text-embedding-ada-002

encoding = tiktoken.get_encoding(ENCODING)
separator_len = len(encoding.encode(SEPARATOR))

f"Context separator contains {separator_len} tokens"

COMPLETIONS_API_PARAMS = {
    # We use temperature of 0.0 because it gives the most predictable, factual answer.
    "temperature": 0.0,
    "max_tokens": 1000,
    "model": COMPLETIONS_MODEL,
}
```

```python
def normalize_text(s, sep_token = " \n "):
    s = re.sub(r'\s+',  ' ', s).strip()
    s = re.sub(r". ,","", s)
    # remove all instances of multiple spaces
    s = s.replace("..",".")
    s = s.replace(". .",".")
    s = s.replace("\n", "")
    s = s.replace("#","")
    s = s.strip()
    return s

def get_embedding(text: str, model=EMBEDDING_MODEL):
    result = openai.Embedding.create(
      model=model,
      input=text
    )
    return result["data"][0]["embedding"]

def vector_similarity(x, y):
    return np.dot(np.array(x), np.array(y))

def order_document_sections_by_query_similarity(query, contexts):
    query_embedding = get_embedding(query)
    # vectory similarity between user input and embedding_dict
    document_similarities = sorted([
        (vector_similarity(query_embedding, doc_embedding), doc_index) for doc_index, doc_embedding in contexts.items()
    ], reverse=True)
    return document_similarities

def compute_doc_emb(df):
  doc_emb = {}
  embedding_file_path = "/content/drive/MyDrive/Hyly/doc_embeddings_greystar.json"
  for idx, r in tqdm(df.iterrows(), total=df.shape[0]):
    try:
      doc_emb[idx] = get_embedding(str(r.content))
    #   doc_emb[idx] = get_embedding(str(idx))

    except Exception as e:
      json.dump(doc_emb, open(embedding_file_path, 'w'))
      print(f"\nEROOR REPORTED AT IDX: {idx}\n{traceback.format_exc()}\n")
      print("waiting for 120 s...")
      time.sleep(120)
      print(f"\nRESUMING FROM IDX: {idx}")
      doc_emb[idx] = get_embedding(str(r.content))
    #   doc_emb[idx] = get_embedding(str(idx))

  json.dump(doc_emb, open(embedding_file_path, 'w'))
  print("Document Embeddings file saved!")
  return doc_emb


def get_answer_from_embedding_dict(user_input, context_embedding, dataframe):

    try:
        most_relevant_document_sections = order_document_sections_by_query_similarity(user_input, context_embedding)[:2]

        for i in most_relevant_document_sections:
            document_section = dataframe.iloc[int(i[1])]
            # print(document_section.completion)

            if isinstance(document_section, pd.DataFrame):
                document_section = document_section.iloc[0]
                    
            # print(f"Question: {user_input}\nAnswer: {document_section['completion']}")
            return str(document_section['completion'])
    except:
        # print("Please call us for more information.")
        return str("Please call us for more information.")
```

```python
def construct_index(directory_path):
    
    max_input_size = 4096 # set maximum input size
    num_outputs = 256 # set number of output tokens
    max_chunk_overlap = 20 # set maximum chunk overlap
    chunk_size_limit = 600 # set chunk size limit

    # define LLM
    llm_predictor = LLMPredictor(llm=OpenAI(temperature=0, model_name="text-ada-001", max_tokens=num_outputs))
    prompt_helper = PromptHelper(max_input_size, num_outputs, max_chunk_overlap, chunk_size_limit=chunk_size_limit)

    documents = SimpleDirectoryReader(directory_path).load_data()
    index = GPTSimpleVectorIndex(
        documents, llm_predictor=llm_predictor, prompt_helper=prompt_helper,
    ) # verbose=True
    index.save_to_disk('/content/drive/MyDrive/Hyly/greystar_gpt_index.json')
    return index

def ask_lenny(json_file):
    index = GPTSimpleVectorIndex.load_from_disk(json_file)
    while True: 
        query = input("How can I help you? ")
        response = index.query(query, response_mode="compact") #  verbose=False
        display(Markdown(f"GreyStar Bot says: <b>{response.response}</b>"))
```
